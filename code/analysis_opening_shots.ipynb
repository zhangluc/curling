```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python

stones = pd.read_csv("../data/2026-main/stones_with_hammer.csv")
ends = pd.read_csv("../data/2026-main/Ends.csv")
games = pd.read_csv("../data/2026-main/Games.csv")
teams = pd.read_csv("../data/2026-main/Teams.csv")

stones.head(10)
```

```python
stones.isna().mean().sort_values(ascending=False)
```

```python
# Enforce deterministic ordering
stones = stones.sort_values(
    ["CompetitionID", "SessionID", "GameID", "EndID", "ShotID"]
)

# Basic integrity checks
print("Unique games:", stones["GameID"].nunique())
print("Unique ends:", stones[["GameID", "EndID"]].drop_duplicates().shape[0])
print("Shots per end (summary):")
stones.groupby(["GameID", "EndID"]).size().describe()
```

```python
stones = stones.merge(
    ends[
        [
            "CompetitionID", "SessionID", "GameID",
            "EndID", "TeamID", "Result", "PowerPlay"
        ]
    ],
    on=["CompetitionID", "SessionID", "GameID", "EndID", "TeamID"],
    how="left"
)

stones[["Result", "PowerPlay"]].isna().mean()
```

```python
# Extract PowerPlay values that are actually present
end_powerplay = (
    stones
    .dropna(subset=["PowerPlay"])
    .groupby(["GameID", "EndID"])["PowerPlay"]
    .first()
    .reset_index()
)

# Ensure all ends are represented (including non–power play ends)
all_ends = stones[["GameID", "EndID"]].drop_duplicates()

end_powerplay = all_ends.merge(
    end_powerplay,
    on=["GameID", "EndID"],
    how="left"
)

# Sanity check
end_powerplay["PowerPlay"].value_counts(dropna=False)
```

```python
# Compare end counts by source for sanity check (19 + 14 + 12 = should be 45 total)
pd.DataFrame({
    "stones": [stones[["GameID", "EndID"]].drop_duplicates().shape[0]],
    "ends":   [ends[["GameID", "EndID"]].drop_duplicates().shape[0]],
    "power":  [end_powerplay.shape[0]]
})
```

```python
stones["team_shot_index"] = stones.groupby(
    ["CompetitionID", "SessionID", "GameID", "EndID", "TeamID"]
).cumcount() + 1

# Quick check
stones.groupby("team_shot_index").size().head(10)
stones.groupby(
    ["CompetitionID", "SessionID", "GameID", "EndID", "TeamID"]
)["team_shot_index"].max().value_counts()
```

```python
example = stones[
    (stones["CompetitionID"] == stones["CompetitionID"].iloc[0]) &
    (stones["GameID"] == stones["GameID"].iloc[0]) &
    (stones["EndID"] == stones["EndID"].iloc[0])
].sort_values("ShotID")

example[["TeamID", "ShotID", "team_shot_index", "Task"]]
```

```python
opening_shots = stones[stones["team_shot_index"] <= 3]

print("Rows in opening_shots:", opening_shots.shape)
opening_shots.head()
```

```python
# Drop any duplicate rows first
opening_shots_unique = opening_shots.drop_duplicates(
    subset=["GameID", "EndID", "TeamID", "team_shot_index"]
)

# Now count opening shots per team per end
team_opening_counts = opening_shots_unique.groupby(
    ["GameID", "EndID", "TeamID"]
)["team_shot_index"].count()

team_opening_counts.value_counts().sort_index()
```

```python
opening_shots_unique[["GameID", "EndID"]].drop_duplicates().shape[0]
```

```python
example = opening_shots[
    (opening_shots["GameID"] == opening_shots["GameID"].iloc[0]) &
    (opening_shots["EndID"] == opening_shots["EndID"].iloc[0]) &
    (opening_shots["TeamID"] == opening_shots["TeamID"].iloc[0])
]

example[["ShotID", "team_shot_index", "Task", "PowerPlay"]]
```

```python
# Sum points for the first 3 shots per team per end
opening_shots_points = opening_shots.groupby(
    ["CompetitionID", "SessionID", "GameID", "EndID", "TeamID"]
)["Points"].sum().reset_index()

opening_shots_points.rename(columns={"Points": "OpeningShotPoints"}, inplace=True)

# Quick check
opening_shots_points.head(10)
```

```python
# Flag strong ends (e.g., ≥ 2 points in first 3 shots)
opening_shots_points["StrongOpening"] = opening_shots_points["OpeningShotPoints"] >= 2

# Count strong vs weak
opening_shots_points["StrongOpening"].value_counts()
```

```python
# Merge PowerPlay info from ends table
# Make sure ends DataFrame has: CompetitionID, SessionID, GameID, EndID, TeamID, PowerPlay
# Here we assume ends DataFrame exists

# Keep only one row per team per end from stones to grab Has_Hammer
hammer_info = stones.drop_duplicates(
    subset=["CompetitionID", "SessionID", "GameID", "EndID", "TeamID"]
)[["CompetitionID", "SessionID", "GameID", "EndID", "TeamID", "Has_Hammer"]]

# Create a column indicating whether this team had the power play
opening_shots_points = opening_shots_points.merge(
    ends[["CompetitionID", "SessionID", "GameID", "EndID", "TeamID", "PowerPlay"]],
    on=["CompetitionID", "SessionID", "GameID", "EndID", "TeamID"],
    how="left"
)

# Merge into opening_shots_points
opening_shots_points = opening_shots_points.merge(
    hammer_info,
    on=["CompetitionID", "SessionID", "GameID", "EndID", "TeamID"],
    how="left"
)

# Boolean: did this team have the power play?
opening_shots_points["IsPowerPlayTeam"] = opening_shots_points["PowerPlay"].notna()
opening_shots_points["HasHammer"] = opening_shots_points["Has_Hammer"] == 1  # adjust depending on CSV
```
