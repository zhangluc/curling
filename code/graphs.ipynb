```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("opening_sequences_baseline.csv")

df = df[df["NumEnds"] >= 10]

top10 = df.sort_values("AvgOpeningPoints", ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(
    data=top10,
    x="AvgOpeningPoints",
    y="OpeningSequence"
)

plt.title("Top 10 Opening Shot Sequences by Average Points")
plt.xlabel("Average Points Scored in End")
plt.ylabel("Opening Shot Sequence")
plt.tight_layout()
plt.show()
```

```python
df_hammer = pd.read_csv("opening_sequences_hammer.csv")
df_hammer = df_hammer[df_hammer["NumEnds"] >= 10]

plt.figure(figsize=(10, 6))
sns.barplot(
    data=df_hammer,
    x="AvgOpeningPoints",
    y="OpeningSequence",
    hue="Has_Hammer"
)

plt.title("Opening Sequence Effectiveness With vs Without Hammer")
plt.xlabel("Average Points")
plt.ylabel("Opening Shot Sequence")
plt.legend(title="Has Hammer")
plt.tight_layout()
plt.show()
```

```python
df_pp = pd.read_csv("opening_sequences_powerplay.csv")
df_pp = df_pp[df_pp["NumEnds"] >= 10]

plt.figure(figsize=(10, 6))
sns.barplot(
    data=df_pp,
    x="AvgOpeningPoints",
    y="OpeningSequence",
    hue="IsPowerPlay"
)

plt.title("Opening Sequences With vs Without Power Play")
plt.xlabel("Average Points")
plt.ylabel("Opening Shot Sequence")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(
    data=df,
    x="NumEnds",
    y="AvgOpeningPoints"
)

plt.axhline(df["AvgOpeningPoints"].mean(), linestyle="--")
plt.title("Sample Size vs Opening Effectiveness")
plt.xlabel("Number of Ends Observed")
plt.ylabel("Average Points")
plt.tight_layout()
plt.show()
```
