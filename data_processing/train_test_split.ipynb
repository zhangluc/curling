```python
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
```

```python
PROJECT_ROOT = Path(__file__).resolve().parents[1]
DATA_DIR = PROJECT_ROOT / "data_processing" / "processed_data"
SAVE_DIR = PROJECT_ROOT / "data_processing" / "train_test_data"
```

```python
ends = pd.read_csv(DATA_DIR / "bayesian_training.csv")
```

```python
test_features = ends[["Has_Hammer", "PowerPlayBool", "EndID", "PrevScoreDiff", "PrevEndDiff", "Result"]].copy()

train_df, test_df = train_test_split(test_features, test_size = 0.2, shuffle = True)
train_df.to_csv(SAVE_DIR / "train_df.csv", index = False)
test_df.to_csv(SAVE_DIR / "test_df.csv", index = False)
```
